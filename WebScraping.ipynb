```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/2024_Summer_Olympics_medal_table"
```

```python
#Write a function to Get and parse html content from a Wikipedia page
my_data = requests.get(url).text
```

```python
scrapes = BeautifulSoup(my_data,'html.parser')
```

```python
scrapes.find('table')
```

```python
scrapes.find_all('table')[3]
```

```python
tables = scrapes.find_all('table')
```

```python
first = tables[0]
first
```

```python
for index,table in enumerate(tables):
    if ('2024 Summer Olympics medal table' in str(table)):
        index_no = index
        print(index_no)
```
